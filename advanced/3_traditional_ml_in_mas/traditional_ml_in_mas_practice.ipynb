### ТЗ №1: Стартап (грантовый проект)

**Заказчик:** Научный руководитель гранта РФРИ

**Задача:** Система для анализа научных статей. Нужно кластеризовать 10 000 статей по темам, не используя LLM (дорого).

**Ограничения:** Бюджет — 0 рублей на API LLM. Нужно сделать на традиционном ML.

**Вопросы к ответу студента:**

- Как будете векторизовать тексты?
- Как оцените качество кластеризации без разметки?

**Пример упрощённого решения:**
```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

vectorizer = TfidfVectorizer(max_features=5000)
X = vectorizer.fit_transform(articles)
kmeans = KMeans(n_clusters=20, random_state=42)
kmeans.fit_predict(X)
# Использовать силуэтный коэффициент для оценки
```

**Пример упрощённого решения:**
```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegressionCV

vectorizer = TfidfVectorizer(ngram_range=(1, 2), max_features=5000)
X = vectorizer.fit_transform(queries)
clf = LogisticRegressionCV(Cs=10, cv=5, max_iter=1000)
clf.fit(X, labels)
# Использовать micro F1 для оценки
```
