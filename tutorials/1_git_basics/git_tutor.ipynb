# Работа с Git и структурой проекта: используем ИИ

## 1. Зачем нужна структура проекта?

Хорошая структура проекта помогает:
- Быстро ориентироваться в коде
- Легко подключать новых разработчиков
- Избегать путаницы с файлами
- Автоматизировать сборку и тестирование

**Минимальная структура любого проекта:**
```
project_name/
├── README.md          # описание проекта
├── .gitignore         # что не попадает в git
├── requirements.txt   # зависимости
└── main.py            # точка входа
```

### Продвинутый вариант (с вложенными папками):
```bash
mkdir -p project/{src,data,tests,logs,config} && cd project && touch README.md .gitignore requirements.txt src/__init__.py src/main.py config/settings.py && echo "# My Project" > README.md
```

### Как попросить ИИ сгенерировать однострочник под ваш проект:
```
Создай однострочник bash для создания структуры Python-проекта.
Нужны папки: src, data, tests, docs, config.
Файлы: README.md, .gitignore, requirements.txt, src/main.py, config/settings.py
Сразу заполни README заголовком.
```

### Результат от ИИ:
```bash
mkdir -p myproject/{src,data,tests,docs,config} && cd myproject && touch README.md .gitignore requirements.txt src/__init__.py src/main.py config/settings.py && echo "# MyProject" > README.md && echo "# Project settings" > config/settings.py
```

## 5. Настройка Git

### Первоначальная настройка (делается один раз):
```bash
git config --global user.name "Ваше Имя"
git config --global user.email "your.email@example.com"
git config --global core.editor "code --wait"  # для VS Code
```

### Проверка настроек:
```bash
git config --list
```

### Настройка SSH для GitHub (чтобы не вводить пароль):
```bash
# Создать ключ
ssh-keygen -t ed25519 -C "your.email@example.com"

# Показать ключ
cat ~/.ssh/id_ed25519.pub

# Добавить ключ в GitHub: Settings -> SSH and GPG keys -> New SSH key
```

## 6. Как быстро изучить команды Git с помощью ИИ

### Промпт для изучения конкретной команды:
```
Объясни команду git merge. Как она работает?
Покажи пример с конфликтом и без конфликта.
Дай шпаргалку с флагами.
```

### Промпт для шпаргалки:
```
Создай шпаргалку по Git для новичка.
Включи:
- основные команды (init, add, commit, push, pull)
- работу с ветками (branch, checkout, merge)
- отмену изменений (reset, revert)
- просмотр истории (log, status, diff)
```

### Минимальная шпаргалка:
```
# Начать проект
git init                    # создать репозиторий
git clone <url>            # скопировать репозиторий

# Ежедневная работа
git status                  # что изменилось
git add <file>              # добавить файл
git add .                   # добавить всё
git commit -m "message"     # закоммитить
git push                    # отправить на GitHub
git pull                    # забрать изменения

# Ветки
git branch                  # список веток
git branch <name>           # создать ветку
git checkout <name>         # переключиться
git checkout -b <name>      # создать и переключиться
git merge <name>            # слить ветку

# Отмена
git reset HEAD <file>       # убрать из stage
git checkout -- <file>      # отменить изменения
git log                     # история коммитов
```

## 7. Работа с ветками

### Зачем нужны ветки:
- Каждый разработчик работает в своей ветке
- Основная ветка (main) всегда рабочая
- Можно делать эксперименты, не ломая основной код

### Основной workflow:
```bash
# 1. Создать ветку для новой фичи
git checkout -b feature/add-api

# 2. Работаем, коммитим
git add .
git commit -m "Add API endpoint"

# 3. Отправляем ветку на GitHub
git push origin feature/add-api

# 4. На GitHub создаём Pull Request
#    (кнопка "Compare & pull request")

# 5. После проверки сливаем в main
git checkout main
git merge feature/add-api
git push origin main

# 6. Удаляем ветку (опционально)
git branch -d feature/add-api
```

### Как ИИ помогает с ветками:
```
Я создал ветку feature/auth, поработал в ней, закоммитил.
Теперь хочу слить её в main. Напиши последовательность команд.
Учти, что в main могли быть изменения, пока я работал.
```

### Решение от ИИ:
```bash
# Переключиться на main и обновить
git checkout main
git pull origin main

# Вернуться в свою ветку
git checkout feature/auth

# Слить main в свою ветку (разрешить конфликты)
git merge main
# если конфликты - исправить, потом git add . && git commit

# Переключиться обратно и слить свою ветку
git checkout main
git merge feature/auth
git push origin main
```
